#Graph

데이터 특징, 분석의 결과, 현재 상태 등에 다양한 분석을 위해 시각화하는 것.

Matplotlib은 특별한 그래프나 디테일한 세팅이 가능.

```python
import numpy as np
from matplotlib import pyplot as plt
```

- Plot

```python
epoch = np.arange(1,100)
#print(epoch)
train_loss = 1.5 / (epoch)
val_loss = .3+1/(epoch)
```

```python
plt.figure(figsize=(10,5))
plt.plot(train_loss, 'r-*')
plt.plot(val_loss,'b-*')
plt.legend(['train','validation'])
plt.title('MNIST')
plt.xlabel('epoch')
plt.ylabel('loss')
plt.show()
```

```python
plt.figure(figsize=(8,8))
t = np.arange(0.,5.,0.2)
plt.plot(t,t,'ro',t,t**2,'bs',t,5*np.sin(t),'g^')
plt.legend(['t', 't**2', '5sin(t)'])
plt.show()
```

- Subplot

```python
names = ['First', 'Business', 'Economy']
prices = [20, 5, 1]
```

```python
plt.figure(figsize=(15,5))

plt.subplot(131) # 1행 3열의 1번째 그래프
plt.bar(names, prices)
plt.title(f'{names[0]}', fontsize=15)

plt.subplot(132) # 1행 3열의 2번째 그래프
plt.scatter(names, prices)
plt.xlabel('class', fontsize=12)

plt.subplot(133) # 1행 3열의 3번째 그래프
plt.plot(names, prices, linewidth=5.0)
plt.suptitle('Ticket Class', fontsize=15)
plt.axis('off')
plt.show()
```
